### 4. Split the sample 80/20 into a training and a test set. Run a regression of `Price` on `Review Scores Rating` and `Neighbourhood `. What is the $R^2$ and RMSE on the test set? What is the coefficient on `Review Scores Rating`? What is the most expensive kind of property you can rent?

```python
from sklearn.model_selection import train_test_split

X = df.loc[:,['Review Scores Rating', 'Neighbourhood '] ]
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=.2, random_state=100)

Z_train = pd.concat([X_train['Review Scores Rating'], pd.get_dummies(X_train['Neighbourhood '], dtype='int')], axis = 1)
Z_test = pd.concat([X_test['Review Scores Rating'], pd.get_dummies(X_test['Neighbourhood '], dtype='int')], axis = 1)

reg = linear_model.LinearRegression(fit_intercept=False).fit(Z_train,y_train)

y_hat = reg.predict(Z_test)
print('RSQ: ', reg.score(Z_test,y_test))
print('RMSE: ', np.sqrt( np.mean( (y_test - y_hat)**2 )))

results = pd.DataFrame({'variable':reg.feature_names_in_, 'coefficient': reg.coef_})
results
```

```python
1.03227 * 100 + 89.42
```

The most expensive property you can rent is a 100-rated property in Manhattan for 193.65 a night

### 5. Split the sample 80/20 into a training and a test set. Run a regression of `Price` on `Review Scores Rating` and `Neighbourhood ` and `Room Type`. What is the $R^2$ and RMSE on the test set? What is the coefficient on `Review Scores Rating`? What is the most expensive kind of property you can rent?

```python
X = df.loc[:,['Review Scores Rating', 'Neighbourhood ', 'Room Type'] ]
y = df['Price']
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=.2, random_state=100)

Z_train = pd.concat(
    [X_train['Review Scores Rating'],
     pd.get_dummies(X_train['Neighbourhood '], dtype='int'),
     pd.get_dummies(X_train['Room Type'], dtype='int')],
    axis = 1)

Z_test = pd.concat(
    [X_test['Review Scores Rating'],
     pd.get_dummies(X_test['Neighbourhood '], dtype='int'),
     pd.get_dummies(X_test['Room Type'], dtype='int')],
    axis = 1)

reg = linear_model.LinearRegression(fit_intercept=False).fit(Z_train,y_train)

y_hat = reg.predict(Z_test)
print('RSQ: ', reg.score(Z_test,y_test))
print('RMSE: ', np.sqrt(np.mean( (y_test - y_hat)**2 )))

results = pd.DataFrame({'variable':reg.feature_names_in_, 'coefficient': reg.coef_})
results
```

### 6. What does the coefficient on `Review Scores Rating` mean if it changes from part 4 to 5? Hint: Think about how multilple linear regression works.

The coefficient changed from 1.03 in part 4 to 0.627 in part 5. During multiple linear regression, variables of interest are initially regressed against all other predictors, followed by saving residuals and regressing them against each other. Part 4's analysis omitted room type information, whereas part 5 included it. This implies that some price variability tied to ratings and neighborhoods is influenced by room types, causing the reduction in the coefficient related to ratings. The other variables become less accurate as predictors by controlling room type because their power is correlated to the room type variable partially.

###   5. Use the `PolynomialFeatures` function from `sklearn` to expand the set of numerical variables you're using in the regression. As you increase the degree of the expansion, how do the $R^2$ and `RMSE` change? At what point does $R^2$ go negative on the test set? For your best model with expanded features, what is the $R^2$ and `RMSE`? How does it compare to your best model from part 4?

```python
from sklearn.preprocessing import PolynomialFeatures

for d in np.arange(1,5):
    expander = PolynomialFeatures(degree=d,include_bias=False)
    
    Z = expander.fit_transform(X_test_n)

    X_test_d = pd.DataFrame(data=Z, columns = expander.get_feature_names_out())
    
    Z = expander.fit_transform(X_train_n)

    X_train_d = pd.DataFrame(data=Z, columns = expander.get_feature_names_out())
    X_train_d.reset_index(drop=True, inplace=True)
    
    X_train_c.reset_index(drop=True, inplace=True)
    
    X_train_all = pd.concat([X_train_d, X_train_c], axis=1, ignore_index=True)
    
    X_test_d.reset_index(drop=True, inplace=True)
    X_test_c.reset_index(drop=True, inplace=True)
    
    X_test = pd.concat([X_test_d, X_test_c], axis=1, ignore_index=True)

    reg = linear_model.LinearRegression().fit(X_train_all,y_train)
    
    print(d, ' Rsq: ', reg.score(X_test,y_test))
    
    y_hat_all = reg.predict(X_test)
    print(d, 'RMSE: ', np.sqrt( np.mean( (y_test - y_hat_all)**2 )))
```

As the degree of expansion increases, initially RSQ increases and RMSE decreases until degree 2. Degree 2 is the best with a RSQ of 0.8058 and a RSME of 0.2051. It is only marginally better than degree 3, however degree 4 is much worse than degree 3 and performance really falls. The best model (degree 2) does slightly better than the joint model in the previous question.
